```python
import json
import pandas as pd
 
# Load the JSON file
# pd.set_option('display.max_rows', None)  # Show all rows
# pd.set_option('display.max_columns', None)  # Show all columns
# pd.set_option('display.width', None)  # Adjust width for wrapping
# pd.set_option('display.max_colwidth', None)  # Adjust column width for long text
with open('CA_category_id.json', 'r') as file:
    data = json.load(file)
 
# Explore the structure to find the key with a list value
top_level_key = None
for key, value in data.items():
    if isinstance(value, list):  # Identify the first key with a list value
        top_level_key = key
        break
 
if top_level_key:
    # Extract the list under the identified key
    items = data[top_level_key]
 
    # Dynamically extract all keys and values
    rows = []
    for item in items:
        row = {}
        for key, value in item.items():
            if isinstance(value, dict):  # If the value is a nested dictionary
                for nested_key, nested_value in value.items():
                    row[f"{key}.{nested_key}"] = nested_value
            else:
                row[key] = value
        rows.append(row)
 
    # Create a DataFrame
    df = pd.DataFrame(rows)
   
 
    # Display the DataFrame
    print(df)
else:
    print("No list found in the JSON structure.")
 
df
 
```

```python
import json
import pandas as pd
 
# Load JSON file
with open('FR_category_id.json', 'r') as file:
    data = json.load(file)
 
# Find the key with the list
items = None
for key, value in data.items():
    if isinstance(value, list):
        items = value
        break
 
if items:
    # Normalize the list into a DataFrame
    df = pd.json_normalize(items)
    print(df)
else:
    print("No list found in the JSON structure.")
```
